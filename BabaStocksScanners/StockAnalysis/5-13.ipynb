```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import yfinance as yf
from pandas_datareader import data as pdr
```

```python
start = '2023-11-12'
end = '2024-02-22'

yf.pdr_override()
df = pdr.get_data_yahoo("TCS.NS", start=start, end=end)
df = df.reset_index()
df = df.drop(['Date', 'High', 'Low', 'Open', 'Volume', 'Adj Close'], axis = 1)
df
```

```python
plt.figure(figsize=(20, 10))
plt.plot(df.Close, 'r', label = 'Close Price')
plt.xlabel("Time (in days)")
plt.ylabel("price")
plt.legend()
plt.show()
```

```python
ma5 = df.Close.rolling(5).mean()
ma13 = df.Close.rolling(13).mean()
```

```python
plt.figure(figsize=(20,10))
plt.plot(df.Close, 'y', label = 'Close')
plt.plot(ma5, 'r', label = '5 days Moving Average')
plt.plot(ma13, 'b', label = '13 days Moving Average')
plt.xlabel("Time")
plt.ylabel("Price")
plt.legend()
```
